```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
#dataset
df = pd.read_csv("SUV_Purchase.csv")
```

## Cleaning and checking data

```python
#checking for missing values
msno.matrix(df)
```

```python
#transporting gender to numbers
def check_gender(dataset):
    output = []
    for data in dataset:
        if data== "Male":
            output.append(0)
        elif data== "Female":
            output.append(1)
    return output

#transforming gender
gender_num = check_gender(list(df.Gender))
df["Gender"]= gender_num
```

## Build the model

### Data preparation

```python
#creating x and y
x= df[["Gender", "Age", "EstimatedSalary"]]
y= df[["Purchased"]]

#transforming df to numpy
x= x.to_numpy()
y= y.to_numpy()
```

### Logic regression model

```python
#splitting data sets
x_train, x_test, y_train, y_test = train_test_split(x, y.ravel(), train_size = 0.7, test_size = 0.3, random_state= 190)

#tranforming and scaling
scaler= StandardScaler()
x_train =scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
#creating and training model
logic=  LogisticRegression()
logic.fit(x_train, y_train)
```

```python
#testing model
logic.score(x_test, y_test)
```

```python
logic.coef_
```

```python
#additional testing
guesses = logic.predict(x_test)
true_labels = y_test

accuracy= accuracy_score(true_labels, guesses)
recall= recall_score(true_labels, guesses)
precision_score= precision_score(true_labels, guesses)
f1_score= f1_score(true_labels, guesses)

print("Model accurace is {}".format(accuracy))
print("Model recall is {}".format(recall))
print("Model precission is {}".format(precision_score))
print("Model F1 score is {}".format(f1_score))
```

```python
#distribution of positive and negative results to check how relevant model actually is
labels_list = list(y_test)
labels_0 = 0
labels_1 = 0
for label in labels_list:
    if label == 0:
        labels_0 += 1
    elif label == 1:
        labels_1 += 1
print("Positive results is {} %".format(labels_1 /len(labels_list)*100))
print("Negative results is {} %".format(labels_0/ len(labels_list)* 100))
```

```python
#puttting results into chart
probab = logic.predict_proba(x_test)
label_1_proba = []
for x in range(len(true_labels)):
    label = true_labels[x]
    probability = probab[x][1]
    if label == 1:
        label_1_proba.append(probability)

label_0_proba = []
for x in range(len(true_labels)):
    label = true_labels[x]
    probability = probab[x][1]
    if label == 0:
        label_0_proba.append(probability)

sns.kdeplot(label_1_proba)
sns.kdeplot(label_0_proba)
plt.legend(["Buy", "Not buy"])
plt.title("Probability vs labels")
```
